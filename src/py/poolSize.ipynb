```python
import math
from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
N = 10000
percent = 20
prevalence = (percent * N) /100.0
p = prevalence/N

def optimal(k):
    return 1/x - np.power((1-p),x) + 1

x = np.linspace(1,50,50).astype(float)
y = optimal(x)
np.amin(y)
np.where(y==np.amin(y))
fig = plt.figure()
plt.plot(x,y,"r")
```

```python
def expected_writes(k):
    number_of_pools = np.ceil(N/k)
    number_of_positive_pools = (1 - np.power((1-p),k)) * number_of_pools
    return number_of_pools + number_of_positive_pools

x = np.linspace(1,50,50).astype(float)
y = expected_writes(x)

np.amin(y)
opt_idx = np.where(y==np.amin(y))
print("optimal poolsize={}".format(x[opt_idx]))
opt_k = y[opt_idx]
print(opt_k)
fig = plt.figure()
new=fig.add_subplot(111)
new.plot(x,y,"black")
new.plot(x[opt_idx]+3,y[opt_idx],"ro")
new.text(x[opt_idx]-13,y[opt_idx]+350,"Optimal Poolsize={}".format("N"))
plt.xlabel("poolsize")
plt.ylabel("E[T]")
```

```python
def expected_actions(k):
    number_of_pools = np.ceil(N/k)
    number_of_positive_pools = (1 - np.power((1-p),k)) * number_of_pools
    return N + number_of_pools + number_of_positive_pools + number_of_positive_pools*k

x = np.linspace(1,50,50).astype(float)
y = expected_actions(x)


np.amin(y)
opt_idx = np.where(y==np.amin(y))
print("optimal poolsize={}".format(x[opt_idx]))
opt_k = y[opt_idx]
print(opt_k)
fig = plt.figure()
new=fig.add_subplot(111)
new.plot(x,y,"black")
new.plot(x[opt_idx],y[opt_idx],"ro")
new.text(x[opt_idx]+2,y[opt_idx],"Optimal Poolsize={}".format(int(x[opt_idx])))
plt.xlabel("poolsize")
plt.ylabel("E[T]")
```
